```
import pandas as pd
```

```
data=pd.read_csv("zoo.data")
data
```

**KFOLD CROSS VALIDATION ( K=5 )**

```
from sklearn.model_selection import KFold
cv = KFold(n_splits=5, random_state=21, shuffle=True)
x_data=data.iloc[:,1:-1].values
y_data=data.iloc[:,-1:].values
for train_index, test_index in cv.split(x_data):
    print("Train Index: ", x_data[train_index], "\n")
    print("Test Index: ", x_data[test_index])

    x_train, x_test, y_train, y_test = x_data[train_index], x_data[test_index], y_data[train_index], y_data[test_index]
```

**" LINEAR KERNEL " FOR MULTICLASS SVM**

```
import matplotlib.pyplot as plt
# Linear kernel 
from sklearn import svm
from sklearn.preprocessing import label_binarize

x_data=data.iloc[:,1:-1].values
y_data=data.iloc[:,-1:].values
# Use label_binarize to be multi-label like settings
Y = label_binarize(y_data, classes=[1,2,3,4,5,6,7])
n_classes = Y.shape[1]

# Split into training and test
from sklearn.model_selection import KFold
cv = KFold(n_splits=5, random_state=21, shuffle=True)

x_train, x_test, y_train, y_test = x_data[train_index], x_data[test_index], Y[train_index], Y[test_index]
# We use OneVsRestClassifier for multi-label prediction
from sklearn.multiclass import OneVsRestClassifier

# Run classifier
classifier = OneVsRestClassifier(svm.LinearSVC(random_state=0))
classifier.fit(x_train, y_train)
y_pred=classifier.predict(x_test)
y_score = classifier.decision_function(x_test)
```

**THE CLASSFICATION ACCURACY  FOR " LINEAR KERNEL "**

---



```
from sklearn.metrics import accuracy_score
print("accuracy:",accuracy_score(y_test,y_pred))
```

**THE AVERAGE PRECISION SCORE FOR " LINEAR KERNEL "**

```

from sklearn.metrics import precision_recall_curve
from sklearn.metrics import average_precision_score

# For each class
precision = dict()
recall = dict()
average_precision = dict()
for i in range(n_classes):
    precision[i], recall[i], _ = precision_recall_curve(y_test[:, i],y_score[:, i])
                                                        
    average_precision[i] = average_precision_score(y_test[:, i], y_score[:, i])

# A "micro-average": quantifying score on all classes jointly
precision["micro"], recall["micro"], _ = precision_recall_curve(y_test.ravel(),y_score.ravel())    
average_precision["micro"] = average_precision_score(y_test, y_score,average="micro")
print('Average precision score, micro-averaged over all classes: {0:0.2f}'.format(average_precision["micro"]))
```

**PRECISION-RECALL CURVE FOR " LINEAR KERNEL " OF MULTICLASS SVM**

```
plt.figure()
plt.step(recall['micro'], precision['micro'], color='b', alpha=0.2,where='post')
plt.fill_between(recall["micro"], precision["micro"], alpha=0.2, color='b')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.title('Average precision score, micro-averaged over all classes: AP={0:0.2f}'.format(average_precision["micro"]))
```

**FROM NEURAL NETWORK**

```
from sklearn.model_selection import train_test_split
x_data=data.iloc[:,1:-1].values
y_data=data.iloc[:,-1:].values

Y = sklearn.preprocessing.label_binarize(y_data, classes=[1, 2, 3, 4, 5, 6, 7])
train_x,test_x,train_y,test_y= train_test_split(x_data, Y, test_size=0.2,random_state=42)

from sklearn.neural_network import MLPClassifier

clf=MLPClassifier(hidden_layer_sizes=(100, ),learning_rate_init=0.005)
clf.fit(train_x,train_y)
predictions=clf.predict(test_x)
```

**THE CLASSFICATION ACCURACY  FOR " NEURAL NETWORK "**

```
from sklearn import metrics
print(metrics.accuracy_score(test_y,predictions))
```
